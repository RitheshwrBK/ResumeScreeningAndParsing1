```python

import re

f = open("/content/drive/MyDrive/Complete_Resume_Screening/Resume_Screening_Project/Colleges.txt", "r")

Lines = f.readlines()

for i in range(len(Lines)):
  Lines[i] = Lines[i].strip()
  if Lines[i].startswith("Indian Institute of Technology"):
    Lines.append(Lines[i])
    x = Lines[i].replace("Indian Institute of Technology", "IIT")
    Lines[i] = x

def generate_ngrams(words, n):
    """
    Generate n-grams from a list of words.
    """
    if len(words) < n:
        return []
    ngrams = zip(*[words[i:] for i in range(n)])
    return [' '.join(ngram) for ngram in ngrams]

def unigram_bigram_trigram_checker(text, phrases):
    """
    Check for the presence of unigrams, bigrams, and trigrams from a string in a list of phrases.
    """
    # Split the text into words
    words = text.split()

    # Generate unigrams, bigrams, and trigrams
    unigrams = generate_ngrams(words, 1)
    bigrams = generate_ngrams(words, 2)
    trigrams = generate_ngrams(words, 3)
    quadgrams = generate_ngrams(words, 4)
    cinquigrams = generate_ngrams(words, 5)
    sixgrams = generate_ngrams(words, 6)
    sevengrams = generate_ngrams(words, 7)
    eightgrams = generate_ngrams(words, 8)
    ninegrams = generate_ngrams(words, 9)
    tengrams = generate_ngrams(words, 10)

    # Combine unigrams, bigrams, and trigrams into one list
    ngrams = unigrams + bigrams + trigrams + quadgrams + cinquigrams + sixgrams + sevengrams + eightgrams + ninegrams + tengrams
    print(ngrams)
    # Check for the presence of ngrams in phrases
    matches = []
    for ngram in ngrams:
        if ngram in phrases:
            matches.append(ngram)

    return matches




def extract_words_after_education(text):
    # Split the text into words
    words = re.findall(r'\b\w+\b', text)
    found_education = False
    education_index = -1
    result_words = []

    # Find the index of "education"
    for i in range(len(words)):
        if words[i].lower() == "education":
            found_education = True
            education_index = i
            break

    # Extract 50 words after "education" if found
    if found_education:
        start_index = education_index + 1
        end_index = min(start_index + 50, len(words))
        result_words = words[start_index:end_index]

    return result_words

# # Example usage
# text = """
# John Doe
# Education:
# Bachelor's degree in Computer Science from Example University.
# Work Experience:
# Software Engineer at ABC Inc.
# Skills:
# Python, Java, JavaScript, AWS.
# """
educations = []



def post_process_tags(tagged_words):
    corrected_tags = []
    i = 0
    while i < len(tagged_words):
        word, tag = tagged_words[i]
        if word.lower() == 'college':
            # Check if "College" is preceded by another word like "Junior"
            if i > 0 and tagged_words[i - 1][0].lower() == 'junior':
                corrected_tags[-1] = (corrected_tags[-1][0], 'ORGANIZATION')
                corrected_tags.append((word, 'ORGANIZATION'))
            # Check if "College" is followed by "of" and then a location
            elif i + 2 < len(tagged_words) and tagged_words[i + 1][0].lower() == 'of' and tagged_words[i + 2][1] == 'LOCATION':
                corrected_tags.append((word, 'ORGANIZATION'))
                corrected_tags.append((tagged_words[i + 1][0], 'ORGANIZATION'))
                corrected_tags.append((tagged_words[i + 2][0], 'ORGANIZATION'))
                i += 2  # Skip the next two words as they are already processed
            else:
                corrected_tags.append((word, tag))
        else:
            corrected_tags.append((word, tag))
        i += 1
    return corrected_tags

# corrected_tags = post_process_tags(clas_text)

def extract_organizations(tagged_words):
    organizations = []
    current_organization = []

    for word, tag in tagged_words:
        if tag == 'ORGANIZATION':
            current_organization.append(word)
        else:
            if len(current_organization) > 1:  # More than one consecutive "ORGANIZATION" tag
                organizations.append(' '.join(current_organization))
            current_organization = []

    # Check if the last collected words form an organization
    if len(current_organization) > 1:
        organizations.append(' '.join(current_organization))

    return organizations

# # Extract and print organizations
# educational_institutions = extract_organizations(corrected_tags)
# for institution in educational_institutions:
#     print(institution)

for string in resume_strs:
  education = []

  content = extract_words_after_education(string)
  content_string = ' '.join(content)
  # print(content)

  matches = unigram_bigram_trigram_checker(string, Lines)
  education.append(matches)
  if matches == []:
    clas_text = st.tag(word_tokenize(content_string)) # clas_text is assigned here, inside the loop
    # print(clas_text)
    corrected_tags = post_process_tags(clas_text)
    print(corrected_tags)
    educational_institutions = extract_organizations(corrected_tags)
    for institution in educational_institutions:
      education.append(institution)
    educations.append(education)
    if len(educations) != (resume+1):
      educations.append(None)

print(educations)
```
